```python
from data.helper import *
```

```python
DF = get_train_df()
```

```python
COLUMNS = [
    "optimiser",
    "features",
    "iterations",
    "pruning",
]
```

```python
for metric in ["n_colours", "mse_loss", "f1_macro"]:
    print(metric)
    df = DF.copy()
    df = df[df.facts == "all"]
    df = df[df.completed]
    Models = {"svr"}
    if metric == "n_colours":
        Models.add("rank-svm")
    df = df[df["optimiser"].isin(Models)]
    # combine rows with same pruning, optimiser and iterations
    df = df.groupby(COLUMNS + ["domain"]).max().reset_index()
    df = df[COLUMNS + ["domain", metric]]
    # pivot table
    df = df.pivot_table(index=COLUMNS, values=metric, columns="domain")
    # make entries int2
    if metric == "n_colours":
        df = df.fillna(0)
        df = df.astype(int)
    display(df)
```
