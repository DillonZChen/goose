```python
for domain in DOMAINS:
    df = DF.copy()
    print(domain)
    df = df[df["domain"] == domain]
    df = df[df["optimiser"].isin({"svr", "rank-svm"})]
    # combine rows with same pruning, optimiser and iterations
    df = df.groupby(["pruning", "optimiser"]).max().reset_index()
    df = df[["pruning", "optimiser", "n_data"]]
    # pivot table
    df = df.pivot_table(index=["pruning", "optimiser"], values="n_data")
    # make entries int2
    df = df.astype(int)
    df = df.fillna(0)
    display(df)
```

```python
for domain in DOMAINS:
    df = DF.copy()
    print(domain)
    df = df[df["domain"] == domain]
    df = df[df["optimiser"].isin({"svr", "rank-svm"})]
    # combine rows with same pruning, optimiser and iterations
    df = df.groupby(["pruning", "optimiser", "iterations"]).max().reset_index()
    df = df[["pruning", "optimiser", "iterations", "n_colours"]]
    # pivot table
    df = df.pivot_table(index=["pruning", "optimiser"], columns="iterations", values="n_colours")
    # make entries int
    df = df.astype(int)
    df = df.fillna(0)
    display(df)
```
